Exploratory Data Analysis

```python
events['event_type'].value_counts(normalize = True)
```

User interaction follows a funnel pattern, with views accounting for ~52% of events and purchases ~5%

```python
events.groupby('event_type')['customer_id'].nunique()
```

~64% of users who view eventually purchase at least once

```python
events['campaign_id'].notnull().mean()
```

All events are associated with a marketing campaign

```python
transactions['discount_applied'].value_counts(normalize = True)
```

```python

df3= pd.merge(
    transactions, customers, how='inner', on='customer_id'
)

df3.groupby('loyalty_tier')['gross_revenue'].mean()
```

Average order value remains similar across loyalty tiers

```python
events['traffic_source']=events['traffic_source'].replace('DIRECT','Direct')
events['traffic_source']=events['traffic_source'].replace('EMAIL','Email')
events['traffic_source']=events['traffic_source'].replace('ORGANIC','Organic')
events['traffic_source']=events['traffic_source'].replace('PAID SEARCH','Paid Search')
events['traffic_source']=events['traffic_source'].replace('SOCIAL','Social')
```

```python
events.groupby('traffic_source')['event_type'].value_counts(normalize=True)
```

Email and Paid search traffic sources show higer purchase converison rates followed by Social compared to Direct and Organic traffic

```python
events.groupby('device_type')['event_type'].value_counts(normalize=True)
```

User behaviour and conversion patterns are consistent across device types

```python
df5 = events['timestamp'].dt.weekday >= 5
events.groupby(df5)['event_type'].value_counts()
```

Weekdays generate higher overall tarffic and transactions  than weekends across all funnel stages
This is expected due to:
- Workday browsing
- Emaid & Paid Search campaign timing
- Higher desktop usage
